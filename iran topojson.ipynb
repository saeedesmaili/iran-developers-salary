```python
import topojson
import json
import numpy as np
from shapely import geometry
```

```python
with open("iran_topo.json") as json_file:
    jdata = json_file.read()
    topoJSON = json.loads(jdata)
```

```python
topoJSON.keys()
```

```python
topoJSON['objects']['iran'].keys()
```

```python
topo_features = topoJSON['objects']['iran']['geometries']
scale = topoJSON['transform']['scale']
translation = topoJSON['transform']['translate']
```

```python
topo_features
```

```python
geoJSON=dict(type= 'FeatureCollection', features = [])

for k, tfeature in enumerate(topo_features):
    geo_feature = dict(id=k, type= "Feature")
    print(geo_feature)
    print(tfeature)
    #geo_feature['properties'] = tfeature['properties']
    geo_feature['geometry'] = geometry(tfeature, topoJSON['arcs'], scale, translation)    
    geoJSON['features'].append(geo_feature)  
```

```python
topo_features = topoJSON['objects']['CHN_adm1']['geometries']
scale = topoJSON['transform']['scale']
translation = topoJSON['transform']['translate']
```

```python
geoJSON=dict(type= 'FeatureCollection', features = [])

for k, tfeature in enumerate(topo_features):
    geo_feature = dict(id=k, type= "Feature")
    print(geo_feature)
    print(tfeature)
    geo_feature['properties'] = tfeature['properties']
    geo_feature['geometry'] = topojson.geometry(tfeature, topoJSON['arcs'], scale, translation)    
    geoJSON['features'].append(geo_feature)  
```
